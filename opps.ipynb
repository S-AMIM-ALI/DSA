```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius


```

```python

radius = float(input("Enter radius: "))
c = Circle(radius)

print("Area:", c.area())
print("Perimeter:", c.perimeter())
```
